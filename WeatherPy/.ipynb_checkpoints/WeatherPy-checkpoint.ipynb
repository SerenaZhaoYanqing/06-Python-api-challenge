```python
#import independecies
import os 
import pandas as pd 
import requests
import json
from api_keys import weather_api_key
from api_keys import g_key
import matplotlib.pyplot as plt
import scipy.stats as st
from scipy.stats import linregress
import time


```

## Generate Cities List

```python
#finding files of csv file where we saved the world cities 
filepath="../Resources/worldcities.csv"
worldcities=pd.read_csv(filepath)

#convering the city_ascii into a list 
city_list=worldcities['city_ascii'].tolist()
#getting 600 city names 
cities=city_list[0:600]
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#creating params holding keys and cities, create url for google api 
#base_url_geocode="https://maps.googleapis.com/maps/api/geocode/json"

#creating partial query url for open weather 
base_url_weather = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url_weather = f"{base_url_weather}appid={weather_api_key}&units={units}&q="

#set up empty list to hold response info 
city_name=[]
lat=[]
lng=[]
max_temp=[]
humidity=[]
cloudiness=[]
wind_speed=[]
country=[]
datetime=[]

#print log starting from 1  
city_index=1

# Loop through the cities and get lat, lng from google api 
#loop through cities and get weather data from open weather api 
for city in cities:
    time.sleep(1) 
    weather=requests.get(query_url_weather + city).json()
    try:
        city_name.append(weather['name'])
        lat.append(weather['coord']["lat"])
        lng.append(weather['coord']["lon"])
        max_temp.append(weather['main']['temp_max'])
        humidity.append(weather['main']['humidity'])
        cloudiness.append(weather['clouds']['all'])
        wind_speed.append(weather['wind']['speed'])
        country.append(weather['sys']['country'])
        datetime.append(weather['dt'])
        print(f"Processing city index{city_index}|{city}")
        
        #increaseing processing city =1
        city_index += 1
    except:
        print("weather data for {city} not found, pass")
    continue
        
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python

#create dataframe to hold above results
weather_df = pd.DataFrame({
    "City": city_name,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed":  wind_speed,
    "Country": country,
    "DateTime":datetime    
})

#show dataframe 
weather_df.head()


#save csv
weather_df.to_csv("city_weather_data.csv")
```

## Inspect the data and remove the cities where the humidity > 100%.

Skip this step if there are no cities that have humidity > 100%. 

```python
#  Get the indices of cities that have humidity over 100%.
# Make a new DataFrame equal to the city data to drop all humidity outliers by index.
# Passing "inplace=False" will make a copy of the city_data DataFrame, which we call "clean_city_data".

#it is easier to just to get all rows of humidity less than 100, and getting clean df (istead of drop)
clean_city_data=weather_df[weather_df.Humidity<=100]
clean_city_data.head()
```

## Linear Regression

```python
northern=clean_city_data[clean_city_data.lat>=0]
southern=clean_city_data[clean_city_data.lat<0]
```

# linear regression analysis: Max temp vs Latitude

In northern hemisphere, there is a negative relation between latitude and max temperature , which means the higher the latitude is, the lower the max temperature will be ( the weather is cooler).
However in southern hemisphere, there is a positive relation relation between latitude and max temperature , which means the higher the latitude is, the higher the max temperature will be ( the weather is warmer).

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x=northern['lat']
y=northern['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot linear regression on top of scatter plot
plt.scatter(x,y)
plt.plot(x,regress_values,"r-")

# Annotate linear regression
plt.annotate(line_eq,(20,80),fontsize=15,color="red")

# Add labels and title to plot
plt.xlabel("Latitude ")
plt.ylabel("Max Temp)")
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.savefig("Linear_N_Temp.png")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x=southern['lat']
y=southern['Max Temp']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

# Plot linear regression on top of scatter plot
plt.scatter(x,y)
plt.plot(x,regress_values,"r-")

# Annotate linear regression
plt.annotate(line_eq,(-15,80),fontsize=15,color="blue")

# Add labels and title to plot
plt.xlabel("Latitude ")
plt.ylabel("Max Temp)")
plt.title('Southern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.savefig("Linear_S_Temp.png")
plt.show()
```
